```python
# Making a line of stars(str)
def print_many (message, n): 
    for i in range (n): # loop that repeats n times
            print (message, end="") # final print

print_many("*",9 )
print ()
print_many("*",5)
print ()
print_many("*",16)
print ()

# Making a box (not hollow, there are all stars within the box)
'''def print_many_box(message, n, h): 
      for i in range (h): 
            print_many(message, n)
            print ()

print_many_box("*", 7, 3) '''

# Making a box that is hollow 
def print_many_hollow_box(message, n, h):
    print_many(message, n)  # Top border
    for i in range(h - 2):  # Middle rows
        print(message, end="")  # Left border
        print(" " * (n - 2), end="")  # Hollow space (there is concatination between the two strings (a hollow string and a symbol str))
        print(message)  # Right border
    print_many(message, n)  # Bottom border

print_many_hollow_box("*", 7, 5)

# Standard variable name is i (especially for loops) 
```

```python
SIZE = 4
# SIZE is a constant: Global variable 

def bar():
    print("#" + 4 * SIZE * "=" + "#")

def mirror():
    # go up 1..SIZE then down SIZE..1
    for line in list(range(1, SIZE + 1)) + list(range(SIZE, 0, -1)):
        left_spaces  = " " * (-2 * line + 2 * SIZE)
        dots         = "." * (4 * line - 4)
        right_spaces = " " * (-2 * line + 2 * SIZE)

        print("|" + left_spaces + "<>" + dots + "<>" + right_spaces + "|")

bar()
mirror()
bar()
```

```python
#With return command
def ftoc(temp):
    tempc = 5.0/9.0* (temp-32)
    return tempc 
# once the function is defined, without the use of print, it returns the calculated value when the user defines the values for the fuction in this case ftoc(temp). This cuts down on the use of print 

ftoc(100)


# But experimentation with printing instead of return will give 

# With print command 
def gtoc(temp):
    tempf = 5.0/9.0* (temp-32)
    print(tempf)

gtoc(900)

# When rationalizing which to use its imporant to understand which one of the two commands is more effective in a given situation. The <return> command gives u a value so u can use it elsewhere. Like for storage, comparison, and or calculation. While <print> only displays the value, with it u cant make further calcuations or comparisons with it. 

# So if u wish to use the value later use <return> if u just want to display it and not use it later use <print> 

# def function (parameter)
# statement 
#...
# return expression 

# Python dosen't require you to declare that your function returns a value, you just return something at its end. 


```
